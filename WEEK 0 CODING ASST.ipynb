```python
from functools import reduce
df=[df_meal,df_center,df_food]
df1 =  pd.merge(df_meal,df_food,on='meal_id')
df=pd.merge(df1,df_center,on='center_id')
df
```

```python
import matplotlib.pyplot as plt
plt.xticks(rotation=70) 
'''Write on comment what you feel this function does'''

#x-axis labels 
plt.xlabel('Food items')

#y-axis labels 
plt.ylabel('Quantity sold')

#plot title 
plt.title('Most popular food')

#save plot 

X=x[["category"]]
xx=X.values.tolist()
xl=[]
for n in xx:
    xl.append(n[0])
Y=x[["num_orders"]]

yy=Y.values.tolist()
yl=[]

for n in yy:
    yl.append(n[0])
xn = range(len(xl))
plt.bar(xn, yl)
plt.xticks(xn, xl)
#plt.bar(xl,yl,color='green')
plt.show()

plt.savefig('./barGraph.png')
```

```python
cp=df[["checkout_price"]]
cp1=cp.values.tolist()

cpp=[]
for n in cp1:
    cpp.append(n[0])
no=df[["num_orders"]]
no1=no.values.tolist()
noo=[]
for n in no1:
    noo.append(n[0])
revenue=[]
for num1, num2 in zip(cpp,noo):
    revenue.append(float(num1) * float(num2))
df['revenue']=revenue    
df.head()
```

```python
df[["month"]]=df[["week"]]//4
df.head()
month=df.month.unique()
mo=pd.pivot_table(data=df,index='month',values='revenue',aggfunc=np.sum)
month_order=np.array(mo)
```

```python
week=df.week.unique()
wo=pd.pivot_table(data=df,index='week',values='revenue',aggfunc=np.sum)
week_order=np.array(wo)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize=(15,5))

ax1.plot(week, week_order, label = 'weekly_graph')
ax1.set_title('Weekly income')
ax1.set_xlabel('week')
ax1.set_ylabel('revenue')

ax2.plot(month, month_order, label = 'monthly_graph')
ax2.set_title('Mothly income')
ax2.set_xlabel('month')
ax2.set_ylabel('revenue')

plt.show()
fig.savefig('./subplot.png')
```
